#### BIVARIATE ANALYSIS ( Is Race/Ehnicity has any impact on student's performance ? )

```python
def race_performace(scores,group_data):
    f,ax=plt.subplots(1,3,figsize=(20,8))
    for i,score in enumerate(scores):
        sns.barplot(
            x=group_data[score].index
            ,y=group_data[score].values
            ,palette='mako',ax=ax[i])
        ax[i].set_title(score,color='#005ce6',size=20)
        for container in ax[i].containers:
            ax[i].bar_label(container,color='black',size=15)


```

```python
group_data2=df.groupby('race_ethnicity')[['average','math_score','reading_score','writing_score']].mean()
print(group_data2)

race_performace(['math_score','reading_score','writing_score'],group_data2)
```

#### Insights 
- Group E students have scored the highest marks. 
- Group A students have scored the lowest marks. 
- Students from a lower Socioeconomic status have a lower avg in all course subjects
